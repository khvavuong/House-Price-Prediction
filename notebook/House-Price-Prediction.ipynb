# House Price Prediction

## 1. Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics  
```

## 2. Data Collection and Pre-Processing

```python
house_price_dataset = pd.read_csv('../data/housing.csv')
```

```python
print(house_price_dataset)
```

```python
house_price_dataset['MEDV']
# MEDV: Median value of owner-occupied homes in $1000s
```

```python
# Check for missing values
house_price_dataset.isnull().sum()
```

```python
# Statistical measures of dataset
house_price_dataset.describe()
```

## 3. Understanding the correlation between various features in dataset

```python
correlation = house_price_dataset.corr()
```

```python
# Constructing a heatmap to understand the correlation
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the date and Target

```python
X = house_price_dataset.drop(['MEDV'], axis = 1)
Y = house_price_dataset['MEDV']

```

```python
print(X)
print(Y)
```

Splitting the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## 4. Model Training

XGBoost Regressor

```python
# Loading the model
model = XGBRegressor()
```

```python
# Training the model with X_train
model.fit(X_train, Y_train)
```

Evaluation

## 5. Prediction on training data

```python
# Accuracy for prediction on training data
training_data_prediction = model.predict(X_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_train, training_data_prediction)

# Mean Absolute data
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("R square error: ", score_1)
print('Mean Absolute Error: ', score_2)
```

Visualizing the actiual Prices and predicted prices

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Price vs Predicted Price')
plt.show()
```

## 6. Prediction on Test Data

```python
# Accuracy for prediction on testing data
test_data_prediction = model.predict(X_test)
```

```python
# R squared error
score_1 = metrics.r2_score(Y_test, test_data_prediction)

# Mean Absolute data
score_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("R square error: ", score_1)
print('Mean Absolute Error: ', score_2)
```
